# Transfer Matrix Calculations

## sip.transfer_matrix(width1, width2, gap, wavelength, device_len, dz)

Transfer matrix calculator of a device which has an architecture determined by parameters.

- **Parameters:**
 -   **width1:** *1d-Array*. Width array of input port of the device. *Length of the array must be equal to `device_len/dz`.*
 -   **width2:** *1d-Array*. Width array of cross port of the device. *Length of the array must be equal to `device_len/dz`.*
 -   **gap:**    *1d-Array*. Gap between two waveguides along the device. *Length of the array must be equal to `device_len/dz`.*
 -   **wavelength:** *1d-Array*. This array could be consist of 1 element or multiple.
 -   **device_len:** *Integer*. In unit of $\mu{m}$.
 -   **dz:** *Integer*. Mesh size along the device. In unit of $\mu{m}$.
- **Returns:**
 - **transfer_matrix:** *4d-Array*. Transfer matrix of the device with varying wavelength and position on device. Length of this array: [len(wavelength), int(device_len/dz), 2, 2]

```python
import siphotonics as sip
import numpy as np

L = 19      # device length
dz = 0.1    # step size 

# Preparing Architecture of the Device
width1 = np.linspace(0.5, 0.5, int(L/dz))
width2 = np.linspace(0.5, 0.5, int(L/dz))
gap = np.linspace(0.2, 0.2, int(L/dz))
wavelength = np.linspace(1.5, 1.6, 3)

# Main Process
tm = sip.transfer_matrix(width1, width2, gap, wavelength, L, dz)
```

## sip.plot_power_output(transfer_matrix, wavelength, device_length, kind='frequency_sweep')
Plots power of the through and cross ports.
- **Parameters:**
 -   **transfer_matrix:** *4d-Array*. Transfer matrix acquired from `sip.transfer_matrix()`.
 -   **wavelength:** *1d-Array*. Wavelength array that is declared above cell.
 -   **device_len:** *Integer*. In unit of $\mu{m}$.
 -   **kind, *optional*:** *String*. Determines kind of the resulting plot. If `kind="device length`, x axis will be device length. If `kind="frequency_sweep`, x axis will be wavelength. 
- **Returns:**
 - **plot:** Plot of the through and cross ports.

```python
sip.plot_power_output(tm, wavelength, L, kind="frequency_sweep")
```

```python
sip.plot_power_output(tm, wavelength, L, kind="device_length")
```
