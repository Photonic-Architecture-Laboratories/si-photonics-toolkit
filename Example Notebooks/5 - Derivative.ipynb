# derivative(data, order=1, step=1)
This function takes an array of data and returns an array of specified ordered derivative of the data up to fourth order.

* It takes **3** arguments:
 * **data**: A data array.
 * **order** *(optional)*: Order of desired derivation. It is *1* by default.
 * **step** *(optional)*: Step size of data array. It is *1* by default.

# Usage

```python
import siphotonics as sip
import numpy as np

t = np.linspace(0, 4*np.pi, 1000) # Step size is '4*np.pi/1000'
y = np.cos(t)                     # Data is a cosine function

first_order = sip.derivative(y, step=4*np.pi/1000)           # -sin
second_order = sip.derivative(y, order=2, step=4*np.pi/1000) # -cos
third_order = sip.derivative(y, 3, 4*np.pi/1000)             # sin
fourth_order = sip.derivative(y, 4, 4*np.pi/1000)            # cos
```

```python
import matplotlib.pyplot as plt

plt.plot(t, y, label="Original cosine", linestyle="solid")
plt.plot(t, first_order,  label="1$^{st}$ order (-sin)", linestyle="dashed")
plt.plot(t, second_order, label="2$^{nd}$ order (-cos)", linestyle="dotted")
plt.plot(t, third_order,  label="3$^{rd}$ order (sin)", linestyle="dashdot")

plt.title("Cosine and Its First Three Derivatives")
plt.legend()
plt.show()
```
